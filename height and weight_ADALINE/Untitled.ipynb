```python
import numpy as np
import pandas as pd
import random
import sklearn
import matplotlib.pyplot as plt
from scipy import stats
from numpy.linalg import inv
from sklearn.neighbors import KNeighborsClassifier
```

```python
N = 200

X = []
Y = []

for i in range(N):
   X.append(np.random.randint(60, 200, dtype=int ))
   s=random.uniform(0.15, 0.4) * X[i]
   Y.append(s)
    
plt.scatter(X,Y,c='red')
```

```python
# m = (X.XT)^-1 *yXT
X=np.array(X)
Y=np.array(Y)
X=X.reshape(200,1)
Y=Y.reshape(200,1)
print(X.shape, Y.shape)
```

```python
m=np.matmul(inv(np.matmul(X.T,X)),np.matmul(X.T,Y))
m
```

```python
m_wrong = random.uniform(0.6, 0.7)
Y_pred_wrong = X * m_wrong
plt.plot(X,Y_pred_wrong,c='blue',lw=1)
plt.scatter(X,Y ,c='red')
```

```python
Y_pred_correct=np.matmul(X,m)
plt.plot(X,Y_pred_correct,c='green',lw=3)
plt.scatter(X,Y ,c='red')
```
